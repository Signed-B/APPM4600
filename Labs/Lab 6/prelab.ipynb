```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.cos(x)
n = np.arange(0, 20)
h = lambda x: 0.01 * (2.0**(-1*x))
```

```python
df1 = (f(np.pi/4 + h(n)) - f(np.pi/4))/h(n)
df2 = (f(np.pi/4 + h(n)) - f(np.pi/4 - h(n)))/(2*h(n))

# plot log_2 of the error
plt.plot(n, np.log2(np.abs(df1 + np.sin(np.pi/4))))
plt.plot(n, np.log2(np.abs(df2 + np.sin(np.pi/4))))
plt.xlabel('n')
plt.ylabel('log_2(error)')
plt.legend(['Forward Difference', 'Central Difference'])
plt.show()
```
