```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.cos(x)
h = 0.01 * (2.0**(-1* np.arange(0, 10)))
```

```python
df1 = (f(np.pi/4 + h) - f(np.pi/4))/h
df2 = (f(np.pi/4 + h) - f(np.pi/4 - h))/(2*h)

# plot log_2 of the error
plt.plot(h, np.log2(np.abs(df1 + np.sin(np.pi/4))))
plt.plot(h, np.log2(np.abs(df2 + np.sin(np.pi/4))))
plt.xlabel('log_2(h)')
plt.ylabel('log_2(error)')
plt.legend(['Forward Difference', 'Central Difference'])
plt.show()
```
