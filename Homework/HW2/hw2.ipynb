```python
# plot a parametric curve
import matplotlib.pyplot as plt
t = np.linspace(0, 2*np.pi, 300)
R = 1.2
dr = 0.1
f = 15
p = 0
x = lambda t: R*(1 + dr*np.sin(f*t + p))*np.cos(t)
y = lambda t: R*(1 + dr*np.sin(f*t + p))*np.sin(t)
plt.plot(x(t), y(t))
plt.title('Parametric curve')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
