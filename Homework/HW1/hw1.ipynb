```python
import numpy as np
import matplotlib.pyplot as plt
```

### Q1

```python
x = np.arange(1.920, 2.080, 0.001)
y1 = x**9 - 18*x**8 + 144*x**7 - 672*x**6 + 2016*x**5 - 4032*x**4 + 5376*x**3 - 4608*x**2 + 2304*x - 512
y2 = (x-2)**9

plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('y1=x^9... and y2=(x-2)^9)')
plt.show()
```

### Q5

```python
x = np.logspace(-16, 0, endpoint=False, num=100)

def f1(x,d):
    return np.cos(x+d) - np.cos(x)
def f2(x,d):
    return -2*np.sin(x+d/2)*np.sin(d/2)

plt.plot(x, np.abs(f1(np.pi, x) - f2(np.pi, x)), label='d=pi')
plt.plot(x, np.abs(f1(1e6, x) - f2(1e6, x)), label='d=10^6')
plt.legend()
plt.xlabel('log(d)')
plt.ylabel('log(error)')
plt.xscale('log')
plt.yscale('log')
plt.title('|mag difference|')
plt.show()
```
